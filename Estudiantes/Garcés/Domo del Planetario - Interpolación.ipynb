```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm as tqmd
from scipy import interpolate
from scipy.interpolate import lagrange

%matplotlib inline
```

```python
# 1. Carguemos y mostremos la imagen
plt.figure(figsize=(8, 6))
imagen = plt.imread('planetario.jpg')
plt.imshow(imagen)
#plt.plot(x_*625, y_*1080, "ro")
plt.show()
```

```python
x_ = np.array([332,335,340,348,357,369,386,409,433,456,486,514,538,569,595,622,666,689,710,725,737,744,745])
y_ = np.array([233,212,192,170,149,128,107,88,71,58,48,42,41,43,49,58,86,106,133,162,191,221,250])
plt.figure(figsize=(8, 6))
plt.imshow(imagen)
plt.plot(x_, y_, "ro")
plt.show()
```

```python
def lagrange_P_n(xi, yi):
    def L(x):
        total = 0.0
        n = len(xi)
        for k in range(n):
            L_k = 1.0
            for i in range(n):
                if i != k:
                    L_k *= (x - xi[i]) / (xi[k] - xi[i])
            total += L_k * yi[k]
        return total

#GRÁFICA: .o.°|*_-:,:|

    x_vals = np.linspace(min(xi), max(xi), 100)
    y_vals = [L(x) for x in x_vals]
    plt.imshow(imagen)
    plt.plot(xi[xi>255], yi[yi<300], '+', label='Datos Originales', color='green', markersize=10)
    plt.plot(x_vals, y_vals, label='Polinomio de Lagrange', color='red')
    plt.title('Interpolación de Lagrange')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
lagrange_P_n(x_ , y_)
```
