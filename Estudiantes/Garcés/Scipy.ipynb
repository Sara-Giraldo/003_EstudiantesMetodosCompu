```python
import numpy as np
import matplotlib.pylab as plt
import scipy.integrate as spi
```

#Odeint

```python
def oscilador(y, t, omega, alpha):
    pos_y = y[0]
    vel_y = y[1]

    dydt = vel_y
    dvdt = -alpha * vel_y - omega * pos_y

    return np.array([dydt, dvdt])
```

```python
y0 = np.array([2, 0])
f = 2
omega = 2 * np.pi * f
alpha = 0.1
t = np.linspace(0, 3, 100)
params = (omega, alpha)

solucion = spi.odeint(oscilador, y0, t, args = params)

pos = solucion[:, 0]
vel = solucion[:, 1]

# GRÁFICA
fig, axs = plt.subplots(2, 1, figsize=(12, 6))
plt.subplots_adjust(hspace=0.4)

# Euler: θ vs t
axs[0].plot(t, pos, color='darkorchid', label="Y")
axs[0].set_title("Y vs. t")
axs[0].set_ylabel("Posición (m)")
axs[0].set_xlabel("Tiempo (t)")
axs[0].grid(True)
axs[0].legend()

# Euler:  ω vs t
axs[1].plot(t, vel, color='sienna', label="V")
axs[1].set_title("V vs. t")
axs[1].set_ylabel("Velocidad (m/s)")
axs[1].set_xlabel("Tiempo (t)")
axs[1].grid(True)
axs[1].legend()
```
