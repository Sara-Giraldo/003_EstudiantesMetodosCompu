```python
import numpy as np

def colisiones(m1, m2, v1, v2):
    colisiones = 0
    epsilon = 1e-10

    if v1 <= v2:
        return 0

    while True:
        # colisión elástica
        v1_nuevo = ((m1 - m2) * v1 + 2 * m2 * v2) / (m1 + m2)
        v2_nuevo = ((m2 - m1) * v2 + 2 * m1 * v1) / (m1 + m2)

        v1, v2 = v1_nuevo, v2_nuevo
        colisiones += 1

        # choca con la pared
        if v1 < 0:
            v1 = -v1
            colisiones += 1

        # Condiciones de parada:
        # 1. no movimiento relativo
        if v1 <= v2:
            break

        # 2. velocidades no cambian
        if abs(v1 - v1_nuevo) < epsilon and abs(v2 - v2_nuevo) < epsilon:
            break

        # 3. demasiadas iteraciones
        if colisiones > 1000000:
            print("Advertencia: Se alcanzó el límite máximo de iteraciones")
            break

    return colisiones

print(colisiones(1, 1, 0, -2))


print(colisiones(1, 10**2, 0, -2))
print(colisiones(1, 10**4 ,0, -2))
print(colisiones(1, 10**6 ,0, -2))
print(colisiones(1, 10**8 ,0, -2))
print(colisiones(1, 10**10 ,0, -2))
```
