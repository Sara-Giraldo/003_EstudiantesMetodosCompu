```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def RK4(a, b, alpha, f, N=10):
    h = (b - a) / N
    w0 = alpha
    t = a
    w = [w0]
    t_values = [a]

    for i in range(1, N + 1):
        t = a + i*h
        k1 = h * f(t, w[i - 1])
        k2 = h * f(t + h / 2, w[i - 1] + k1 / 2)
        k3 = h * f(t + h / 2, w[i - 1] + k2 / 2)
        k4 = h * f(t + h, w[i - 1] + k3)
        wi = w[i - 1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        w.append(wi)
        t_values.append(t)
    return (t_values, w)
```

```python
lambda_ = 0.5
A0 = 1
a = 0
b = 20

def g(t, y):
    return -lambda_ * y


t_vals, N_vals = RK4(a, b, A0, g, 100)
t_exact = np.linspace(a, b, 1000)
N_exact = A0 * np.exp(-lambda_ * t_exact)
plt.plot(t_vals, N_vals, "o-", label="RK4", markersize=3)
plt.plot(t_exact, N_exact, "--", label="Solución exacta")
plt.xlabel("Tiempo")
plt.ylabel("N")
plt.title("Decaimiento radiactivo")
plt.legend()
plt.grid()
plt.show()
```
