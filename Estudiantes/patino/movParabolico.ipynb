```python
import sympy
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

g = 9.8  # Aceleracion de la gracedad m/s2
vo = 50   #rapidez incial
b = 10  # parametro de friccion
theta = np.radians(45) #theta en radianes

#calculo de las velocidades en "x" y "y"
Vx = vo*np.cos(theta) #velocidad en x
Vy = vo*np.sin(theta) #velocidad en y

def poscicionX(t):
  x=(Vx / b)*(1-np.exp(-b*t))
  return x

def poscicionY(t):
  y= (1/b)*((g/b)+Vy)*(1-np.exp(-b*t)) - (g/b)*t
  return y

def f(R,vox, voy, g, b ):
  return (g/b + voy)*R/vox + (g/b**2)*np.log(1-(R*b)/vox)


T=np.linspace(0,4,1000)

root = optimize.root_scalar(poscicionY, x0 = 3.6)
Raiz = root.root

T=np.linspace(0,Raiz,1000)
print(Raiz)

X=poscicionX(T)
Y=poscicionY(T)
R=poscicionX(Raiz)

plt.plot(X,Y)
plt.grid()
plt.xlabel("x (m)", fontsize=14)
plt.ylabel("y (m)", fontsize=14)
plt.title("x vs y", fontsize=14)
plt.show()
print(R)
plt.plot(T,Y)
plt.grid()
plt.xlabel("T (s)", fontsize=14)
plt.ylabel("y (m)", fontsize=14)
plt.title("T vs y", fontsize=14)
plt.show()
print(R)
```
